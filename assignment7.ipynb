- The expected value of rolling two dice and adding the result together gets more complex. There's 36 possible scenarios, since for each of the six rolls possible from die 1, there are 6 possible rolls from die 2. Going through the possibilities, there's one case where the die add to 2 (1 and 1), two cases where the die add to 3 (1 and 2, and 2 and 1), three cases where the die add to 4 (1 and 3, 3 and 1, 2 and 2), and so on. The expected value ca be calculated by multiplying the sum of the die rolls by 1 over the number of possible scenarios, i.e.: --> (1/36 * 2) + (2/36 * 3) + (3/36 * 4) + (4/36 * 5) + (5/36 * 6) + (6/36 * 7) + (5/36 * 8) + (4/36 * 9) + (3/36 * 10) + (2/36 * 11) + (1/36 * 12) = 7. So the expected value is 7.

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

mb = pd.read_csv('metabric.csv')

plt.hist(mb['Mutation Count'], bins = 50)
```

### Likelihood

The likelihood would look like:    $\prod_{i=1}^n \frac{ \lambda^{y_i} e^{-\lambda} }{y_i!}$

### Maximizing the likelihood

Find log-likelihood

$$
l(\lambda) = log(\mathscr{L}(\lambda)) = \sum_{i=1}^{n} log(\lambda^{y_i}) + log(e^{-\lambda}) - log(y_i!) = \sum_{i=1}^{n} {y_i}log(\lambda) - \lambda - log(y_i!)
$$

Then differentiate and set to 0 to maximize

$$
\frac{dl}{d\lambda} (\sum_{i=1}^{n} {y_i}log(\lambda) - \lambda - log(y_i!)) = \sum_{i=1}^{n} (\frac{y_i}{\lambda} - 1) = -n + \frac{1}{\lambda} \sum_{i=1}^{n} y_i = 0
$$

Solve for lambda

$$
\lambda = \frac{1}{n} \sum_{i=1}^{n} y_i 
$$

Looks like the MLE is the sample mean as predicted.

### Plotting the distribution of my model and comparing it to the KDE

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
# AI assisted because I didn't know how to sample from Poisson

lambda_param = np.mean(mb['Mutation Count'])
x = np.arange(0, 45)
pmf = poisson.pmf(x, lambda_param)

fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(14,5))

ax[0].plot(x, pmf, label='Modeled Poisson PMF')
sns.kdeplot(mb['Mutation Count'], label='KDE from Data', ax=ax[0])
ax[0].legend()
ax[0].set_title('Poisson Density vs. KDE from Data')


samples = np.random.poisson(lambda_param, size=len(mb))
sns.ecdfplot(data=samples, linewidth=2, label='Modeled Poisson ECDF', ax=ax[1])
sns.ecdfplot(data=mb['Mutation Count'], linewidth=2, label='ECDF from Data', ax=ax[1])
ax[1].legend(loc='lower right')
ax[1].set_title('Poisson ECDF vs. ECDF from Data')
```

It appears that the peak of our modeled distribution is pretty close to being in the right location, but still not quite in the correct spot. It is also peaked too high when compared to the real data. 

### Bootstrapping density/distribution of my parameter $\lambda$

```python
# THIS CODE IS TAKEN FROM PROF. TERRY's VERSION

# Estimation:
# Terry found mu and sigma for a normal distribution
# I have a poisson distribution, but lambda_hat is effectively the same as mu_hat
def mle(y):
    lambda_hat = np.mean(y)
    return lambda_hat

# Bootstrapping:
S = 5000
sample = [ mle(mb['Mutation Count'].sample(frac=1.0,replace=True)) for s in range(S)]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.kdeplot(x=sample, ax=axes[0], label='kde')
axes[0].set_title('Sampling Density, $\\hat{\\lambda}$')
axes[0].set_xlabel('$\\hat{\\lambda}$')

sns.ecdfplot(x=sample, ax=axes[1], label='ecdf')
axes[1].set_title('Sampling Distribution, $\\hat{\\lambda}$')
axes[1].set_xlabel('$\\hat{\\lambda}$')
plt.tight_layout()
plt.show()
```
